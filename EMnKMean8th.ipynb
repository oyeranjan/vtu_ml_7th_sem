```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
from sklearn.mixture import GaussianMixture

data_set = load_iris()
X = pd.DataFrame(data_set.data)
X.columns = ['Sepal_Length', 'Sepal_Width', 'Petal_Length', 'Petal_Width']
y = pd.DataFrame(data_set.target)
y.columns = ['Targets']

plt.figure(figsize=(14, 14))
colormap = np.array(['red', 'green', 'blue'])

# REAL PLOT
plt.subplot(2,2, 1)
plt.title('Real')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.scatter(X.Petal_Length, X.Petal_Width, c=colormap[y.Targets], s=20)

# K-PLOT
plt.subplot(2, 2, 2)
model = KMeans(n_clusters=3)
y=model.fit(X)
plt.title('KMeans')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.scatter(X.Petal_Length, X.Petal_Width, c=colormap[y.labels_], s=20)

# GMM PLOT
gmm = GaussianMixture(n_components=3)
gmm.fit(X)
y_cluster_gmm = gmm.predict(X)
plt.subplot(2,2, 3)
plt.scatter(X.Petal_Length, X.Petal_Width, c=colormap[y_cluster_gmm], s=20)
plt.title('GMM Classification')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
```
